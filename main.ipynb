```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./Billionaires Statistics Dataset.csv')
df.head()
```

```python
df.columns
```

```python
df=df.rename(columns={'finalWorth':'final_Worth','personName':'person_Name','countryOfCitizenship':'country_Of_Citizenship','selfMade':'self_Made','birthDate':'birth_Date','lastName':'last_Name','firstName':'first_Name','residenceStateRegion':'residence_State_Region','birthYear':'birth_Year','birthMonth':'birth_Month','birthDay':'birth_Day'})
df=df.rename(columns=str.lower)
```

```python
df.columns
```

```python
totalNa = df.isnull().sum()
print(totalNa)
```

```python
df['gdp_country'] = df['gdp_country'].str.replace('$','', regex=False)
df['gdp_country'] = pd.to_numeric(df['gdp_country'].str.replace(',', '', regex=False), errors='coerce')
```

```python
df.fillna(df.mean, inplace=True)
```

```python
df.dtypes
```

```python
totalNa = df.isnull().sum()
print(totalNa)
```

```python
df = df.drop(columns=['birth_day','birth_year','birth_month','birth_date','first_name','last_name','country_of_citizenship','date','category'], axis=1)
```

```python
df['family_share'] = df['person_name'].str.contains('& family', case=False, na=False)
df['person_name'] = df['person_name'].str.replace(' & family', '', regex=False)
```

```python
# converting into Billions
df['final_worth'] = df['final_worth'] / 1000
```

```python
df.describe()
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce') 
# Create subplots for age and final_worth histograms
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Age Distribution Histogram
ax1.hist(df['age'].dropna(), bins=range(30, 80, 5), edgecolor='black', color='skyblue', alpha=0.7)
ax1.axvline(df['age'].mean(), color='red', linestyle='dashed', linewidth=1, label='Mean')
ax1.axvline(df['age'].median(), color='blue', linestyle='dashed', linewidth=1, label='Median')
ax1.set_title('Age Distribution of Billionaires')
ax1.set_xlabel('Age')
ax1.set_ylabel('Frequency')
ax1.legend()

# Worth Distribution Histogram
ax2.hist(df['final_worth'], bins=range(1000, 6000, 500), edgecolor='black', color='skyblue', alpha=0.7)
ax2.axvline(df['final_worth'].mean(), color='red', linestyle='dashed', linewidth=1, label='Mean')
ax2.axvline(df['final_worth'].median(), color='blue', linestyle='dashed', linewidth=1, label='Median')
ax2.set_title('Worth Distribution of Billionaires')
ax2.set_xlabel('Worth')
ax2.set_ylabel('Frequency')
ax2.legend()

plt.show()
```

```python
# Calculate gender percentages
gender_percentages = df['gender'].value_counts(normalize=True) * 100

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(gender_percentages, labels=gender_percentages.index, autopct='%0.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.legend(gender_percentages.index, title='Gender', loc='upper right')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df['rank'], df['final_worth'], linewidth=1)
plt.scatter(df['rank'], df['final_worth'], c=df['gender'].map({'Male': 'blue', 'Female': 'red'}), s=50)
plt.title('Line Chart: Rank vs. Final Worth (Colored by Gender)')
plt.xlabel('Rank')
plt.ylabel('Final Worth')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df['self_made'], df['final_worth'], color=['blue', 'orange'])
plt.title('Self-Made vs. Inherited Billionaires')
plt.xlabel('Billionaires')
plt.ylabel('Mean Net Worth (in billions)')
plt.xticks(range(len(df['self_made'])), labels=df['self_made'])
plt.show()
```

```python
# Group by 'industries' and calculate total and average wealth
industry_wealth = df.groupby('industries').agg(avg_wealth=('final_worth', 'mean'),
                                               total_wealth=('final_worth', 'sum')).reset_index()

# Plot average wealth by industry
plt.figure(figsize=(10, 6))
industry_wealth = industry_wealth.sort_values(by='avg_wealth', ascending=False)
plt.barh(industry_wealth['industries'], industry_wealth['avg_wealth'], color='skyblue')
plt.xlabel('Average Wealth (Billions)')
plt.ylabel('Industry')
plt.title('Average Wealth by Industry')
plt.gca().invert_yaxis()  # Invert the y-axis
plt.show()

# Plot total wealth by industry
plt.figure(figsize=(10, 6))
industry_wealth = industry_wealth.sort_values(by='total_wealth', ascending=False)
plt.barh(industry_wealth['industries'], industry_wealth['total_wealth'], color='skyblue')
plt.xlabel('Total Wealth (Billions)')
plt.ylabel('Industry')
plt.title('Total Wealth by Industry')
plt.gca().invert_yaxis()  # Invert the y-axis
plt.show()
```

```python
# Bin 'age' into groups of 10 and calculate average wealth and family share counts
bins = list(range(20, 101, 10))
df['age_bin'] = pd.cut(df['age'], bins, right=False)
age_bin_data = df.groupby('age_bin').agg(avg_wealth=('final_worth', 'mean'),
                                         count_family_share=('family_share', 'sum')).reset_index()

# Create a histogram for average wealth by age group
plt.figure(figsize=(10, 6))
plt.bar(age_bin_data['age_bin'].astype(str), age_bin_data['avg_wealth'], color='pink', edgecolor='black', alpha=0.8)
plt.xlabel('Age Group')
plt.ylabel('Average Wealth (Billions)')
plt.title('Average Wealth by Age Group (Bins of 10)')
plt.xticks(rotation=45, ha='right')
plt.show()

# Create a histogram for the number of family shares by age group
plt.figure(figsize=(10, 6))
plt.bar(age_bin_data['age_bin'].astype(str), age_bin_data['count_family_share'], color='lightyellow', edgecolor='black', alpha=0.8)
plt.xlabel('Age Group')
plt.ylabel('Number of Family Share')
plt.title('Number of Family Share by Age Group (Bins of 10)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Identify the top 10 countries with the highest number of billionaires
top_countries = df['country'].value_counts().nlargest(10).reset_index()
top_countries.columns = ['country', 'count']

# Create a bar chart for the top 10 countries
plt.figure(figsize=(10, 6))
plt.bar(top_countries['country'], top_countries['count'], color='coral')
plt.xlabel('Country')
plt.ylabel('Number of Billionaires')
plt.title('Top 10 Countries with Most Billionaires')
plt.xticks(rotation=45, ha='right')
plt.show()

# Filter the dataset for billionaires from the top 3 countries
top_3_countries = top_countries['country'].head(3).to_list()
top_countries_billionaires = df[df['country'].isin(top_3_countries)]

# Create a stacked bar chart for the top 3 countries by industry
plt.figure(figsize=(10, 6))
top_countries_billionaires.groupby(['industries', 'country']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Top 3 Countries Contribution to Each Industry')
plt.xlabel('Industry')
plt.ylabel('Number of Billionaires')
plt.legend(title='Country')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Calculate the number of billionaires and average age for each industry
industry_stats = df.groupby('industries').agg(num_billionaires=('industries', 'count'),
                                              avg_age=('age', 'mean')).reset_index()

# Create a combined plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the number of billionaires
ax1.bar(industry_stats['industries'], industry_stats['num_billionaires'], color='aquamarine', alpha=0.7, label='Number of Billionaires')
ax1.set_xlabel('Industry')
ax1.set_ylabel('Number of Billionaires')
ax1.tick_params(axis='x', rotation=45, labelright=True)
ax1.legend(loc='upper left')

# Create a secondary y-axis for average age
ax2 = ax1.twinx()
ax2.plot(industry_stats['industries'], industry_stats['avg_age'], color='red', label='Average Age', marker='o')
ax2.set_ylabel('Average Age')
ax2.legend(loc='upper right')

# Set the title and labels
plt.title('Billionaires Statistics by Industry')
plt.grid()
plt.show()
```
